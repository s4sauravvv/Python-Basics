```python
#1. Write a query to display the names (first_name, last_name) using alias name "First Name", "Last Name" 
```

```python
sql="select first_name 'First Name', last_name 'Last Name' from Employee"
result=cursor.execute(sql)
for row in result:
    print(row)
```

```python
#2. Write a query to get unique department ID from employee table
```

```python
sql="select DISTINCT department_id from Employee" #Dictinct use for unique element
result=cursor.execute(sql)
for row in result:
    print(row)
```

```python
#3. Write a query to get all employee details from the employee table order by first name, descending. 
```

```python
sql="select   first_name from Employee order by first_name desc "
result=cursor.execute(sql)
for row in result:
    print(row)
```
